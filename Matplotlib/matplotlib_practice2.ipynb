```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
import numpy as np
x = np.arange(10)
y = x**2
    
```

```python
y
```

```python
fig = plt.figure()
axes = fig.add_axes([0,0,1,1])
axes.plot(x,y,label = "x,y")
axes.legend(loc=1)
plt.xlabel("x label")
plt.ylabel("y label")
plt.title("fig")

```

```python
fig = plt.figure()
axes1 = fig.add_axes([0,0,1,1])
axes2 = fig.add_axes([0.1,0.6,0.2,0.2])
axes1.plot(x, y)
axes2.plot(x,y)
```

```python
fig, axes = plt.subplots (nrows = 3, ncols =1)
for ax in axes:
    ax.plot(x,y)
fig
```
