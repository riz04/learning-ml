```python
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
from sklearn.datasets import make_blobs
```

```python
data = make_blobs(n_samples=1000, n_features=2, 
                  centers= 4 , cluster_std= 3.0, random_state= 42)
```

```python
data
```

```python
data[0] , len(data[0])
```

```python
data[0].shape
```

```python
plt.scatter ( x = data[0][: , 0],
             y = data[0][: , 1] , c = data[1] )
```

```python
from sklearn.cluster import KMeans
```

```python
kmeans = KMeans(n_clusters = 4)
```

```python
kmeans.fit(data[0])
```

```python
cent = kmeans.cluster_centers_
cent
```

```python
kmeans.labels_
```

```python
fig , (ax1 , ax2) = plt.subplots (nrows = 1 , ncols = 2, sharey = True , figsize = (10,6))

ax1.scatter(x = data[0][: , 0], y = data[0][: , 1], c = kmeans.labels_)
ax1.scatter(x = cent[:,0],y = cent[:,1] , s = 100 , c = "black")
        
ax2.scatter(x = data[0][: , 0], y = data[0][: , 1], c = data[1])
```

```python
sum_sq = { }
for k in range(1,10):
    kmeans = KMeans(n_clusters = k).fit(data[0])
    sum_sq[k] = kmeans.inertia_
```

```python
sum_sq
```

```python
plt.plot(list (sum_sq.keys()),
        list (sum_sq.values()))
```
