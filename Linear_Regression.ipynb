```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
df = pd.read_csv("insurance.csv")
```

```python
df.head()
```

```python
def map_smoking (column):
    mapped = []
    for i in column:
        if i == "yes":
            mapped.append(1)
        else:
            mapped.append(0)
    return mapped
            
```

```python
new_col = map_smoking(df["smoker"])
```

```python
new_col
```

```python
df["smoker_norm"] = new_col
df.head()
    
```

```python
def map_obese (column):
    mapped = []
    for i in column:
        if i > 30:
            mapped.append(1)
        else:
            mapped.append(0)
    return mapped

df["obese"] = map_obese(df["bmi"])
df.head()
```

```python
def map_sex(column):
    mapped = []
    for i in column:
        if i == "male":
            mapped.append(1)
        else:
            mapped.append(0)
    return mapped

df["sex_norm"] = map_sex(df["sex"])
df.head()
```

```python
x = df[['age', 'bmi', 'children', 'smoker_norm', 'obese', 'sex_norm']]
y = df[["charges"]]
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y, test_size = 0.33, random_state = 101) 
```

```python
x_train.head()
```

```python
from sklearn import linear_model
```

```python
lm = linear_model.LinearRegression()
```

```python
lm.fit(x_train,y_train)
```

```python
lm.intercept_
```

```python
lm.coef_
```

```python
pred = lm.predict(x_test)
```

```python
plt.scatter(y_test,pred)
```
